```python
#Гистограмма для числовой переменной age:

sns.histplot(df['age'], kde=True)
plt.title('Распределение age')
plt.xlabel('age')
plt.ylabel('Частота')
plt.show()
```

```python
#Гистограмма для категориальной переменной capital-loss:

sns.histplot(df['capital-loss'], kde=True)
plt.title('Распределение capital-loss')
plt.xlabel('capital-loss')
plt.ylabel('Частота')
plt.show()
```

```python
#Боксплот для числовой переменной hours-per-week:
sns.boxplot(x=df['hours-per-week'])
plt.title('Boxplot hours-per-week')
plt.xlabel('hours-per-week')
plt.show()
```

⏰
Первый график (Распределение возраста): График показывает распределение возраста в выборке. Это гистограмма с линией плотности распределения (KDE). Распределение напоминает левосторонне асимметричное, где наибольшая частота наблюдается в диапазоне от 20 до 40 лет, и затем частота постепенно снижается после 50 лет. Визуально это близко к нормальному распределению, но с заметным "хвостом" справа, что свидетельствует о правосторонней асимметрии.
Второй график (Распределение capital-loss): Этот график имеет крайне высокую плотность на значении около 0 с резким спадом. Большинство наблюдений связано с нулевыми значениями потерь капитала, что создаёт сильно скошенное распределение с длинным правым хвостом, указывая на редкие, но значительные потери капитала у некоторых респондентов. Это распределение далёко от нормального и скорее демонстрирует крайне скошенное распределение.
Третий график (Boxplot часов работы в неделю): Боксплот для переменной "hours-per-week" показывает медиану и межквартильный размах, а также многочисленные выбросы с обеих сторон. Основная масса данных сосредоточена вокруг значения 40 часов в неделю, что можно считать типичной продолжительностью рабочей недели. Существуют значительные выбросы как в сторону меньших значений (от 0 до 40), так и в сторону больших значений (от 40 до 100).
